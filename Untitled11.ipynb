```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(15)

dataSet = np.random.normal(100, 25, 200)

print(dataSet)

figure = plt.figure(figsize =(10, 8))
plt.boxplot(dataSet)
plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plot # List of values

data_value = [(3, 5),

(25, 30),

(7, 4),

(14, 15)]

dataFrame = pd.DataFrame(data = data_value, columns = ['x_axis', 'y_axis']); # Draw a scatter plot using pandas

dataFrame.plot.scatter(x = 'x_axis', y = 'y_axis', title = "Scatter plot pandas variable for X and Y axis");
```

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7]

plt.hist(data, bins=5, color='skyblue', edgecolor='black')

plt.xlabel('Data')
plt.ylabel('Frequency')
plt.title('Histogram Example')

plt.show()
```
